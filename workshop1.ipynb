```python
import numpy as np

x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])
```

```python
 #1. Add
print("x + y:\n", x + y)
```

```python
# 2. Subtract
print("\nx - y:\n", x - y)
```

```python
# 3. Multiply by integer
print("\nx * 3:\n", x * 3)
```

```python
# 4. Square elements
print("\nx squared:\n", x**2)
```

```python
# 5. Dot products
print("\ndot(v, w):", np.dot(v, w))
print("x dot v:", x.dot(v))
print("x dot y:\n", x.dot(y))
```

```python
# 6. Concatenate x & y (rows), v & w (columns)
print("\nConcatenate x,y along rows:\n", np.concatenate((x,y), axis=0))
concat_vw = np.vstack((v, w)).T
print("\nConcatenate v & w as columns:\n", concat_vw)
```

```python
# 7. Why x and v cannot concatenate?
try:
    print(np.concatenate((x, v)))
except Exception as e:
    print("\nError concatenating x and v:\n", e)
    print("Reason: x is 2x2, v is 1D length-2. Their shapes do not match on axis 0.")
```

```python
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
```

```python
# 1. A * A^-1 = I
A_inv = np.linalg.inv(A)
print("A inverse:\n", A_inv)
print("\nA * A^-1:\n", A.dot(A_inv))
```

```python
# 2. AB != BA
AB = A.dot(B)
BA = B.dot(A)
print("\nAB:\n", AB)
print("\nBA:\n", BA)
print("AB == BA? ->", np.array_equal(AB, BA))
```

```python
# 3. (AB)^T = B^T A^T
lhs = (A.dot(B)).T
rhs = B.T.dot(A.T)
print("\n(AB)^T:\n", lhs)
print("\nB^T A^T:\n", rhs)
print("Equal? ->", np.array_equal(lhs, rhs))
```

```python
# 4. Solve system using matrix form
M = np.array([[2,-3,1],
              [1,-1,2],
              [3,1,-1]])

rhs = np.array([-1,-3,9])

solution = np.linalg.solve(M, rhs)
print("\nSolution using solve():\n", solution)
```

```python
# 5. Solve using inverse
solution_inv = np.linalg.inv(M).dot(rhs)
print("\nSolution using inverse:\n", solution_inv)
```
